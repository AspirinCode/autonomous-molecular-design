```python
from deepchem.utils.save import load_from_disk
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
###Display logD histogram
dataset_file= "./enamineSubset10KGroundTruth.csv"
train_dataset_file= "./lipophilicity.csv"

dataset = load_from_disk(dataset_file)
train_dataset = load_from_disk(train_dataset_file)

logDs = np.array(dataset["logD"])
train_logDs = np.array(train_dataset["exp"])

n, bins, patches = plt.hist(logDs, 50, facecolor='green', alpha=.5, label="REAL", density=True)
n, bins, patches = plt.hist(train_logDs, 50, facecolor='blue', alpha=0.5, label="TRAIN", density=True)

plt.xlabel('Ground truth train and predicted logD')
plt.ylabel('Number of compounds')
plt.title(r'Histogram of train and GT predicted logDs')
plt.grid(True)
plt.legend()
plt.yticks([])
plt.show()
```

```python
###Display logD histogram for both subsets
dataset_file= "./enamineSubset10KGroundTruth.csv"
train_dataset_file= "./lipophilicity.csv"
hundredK_file = "./enamineSubset100KGroundTruth.csv"

dataset10 = load_from_disk(dataset_file)
dataset100 = load_from_disk(hundredK_file)
train_dataset = load_from_disk(train_dataset_file)

logDs10 = np.array(dataset10["logD"])
logDs100 = np.array(dataset100["logD"])
train_logDs = np.array(train_dataset["exp"])

plt.hist(logDs10, 50, facecolor='green', alpha=.5, label="10K Enamine Subset", density=True)
plt.hist(logDs100, 50, facecolor='red', alpha=.5, label="100K Enamine Subset", density=True)
plt.hist(train_logDs, 50, facecolor='blue', alpha=.5, label="Training Data", density=True)

plt.xlabel('Partition coefficient logD')
plt.ylabel('Number of compounds')
plt.title(r'Training and ground truth model predicted logDs')
plt.grid(True)
plt.legend()
plt.yticks([])
plt.show()
```
